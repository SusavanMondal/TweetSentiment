```python
import lxml
df['text']=df['text'].apply(lambda x: x.lower())
df['text']=df['text'].apply(lambda x: ps.remove_accented_chars(x))
# df['text']=df['text'].apply(lambda x: ps.remove_common_words(x))
df['text']=df['text'].apply(lambda x: ps.remove_dups_char(x))
df['text']=df['text'].apply(lambda x: ps.remove_emails(x))
df['text']=df['text'].apply(lambda x: ps.remove_html_tags(x))
# df['text']=df['text'].apply(lambda x: ps.remove_rarewords(x))
df['text']=df['text'].apply(lambda x: ps.remove_rt(x))
df['text']=df['text'].apply(lambda x: ps.remove_special_chars(x))
# df['text']=df['text'].apply(lambda x: ps.remove_stopwords(x))
df['text']=df['text'].apply(lambda x: ps.remove_urls(x))








```

```python
#train test split
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test=train_test_split(df['text'], df['sentiment'], test_size=0.2, random_state=0)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
import joblib

# Example of fitting a model with text data
vectorizer = TfidfVectorizer()
X_train_transformed = vectorizer.fit_transform(X_train)  # X_train should be a list of text samples

model = RandomForestClassifier()
model.fit(X_train_transformed, Y_train)  # y_train should be your labels

# Save the model and vectorizer with compression
joblib.dump((vectorizer, model), 'twitter_model_compressed.pkl', compress=3)
```
